```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [5, 5]
```

```python
df = pd.read_csv("td_data.csv")
print(df.to_string())
```

```python
ax = df.plot.bar(x="Project Name", y="Z-value", rot=90, color="b")
plt.axhline(y=2.58, color='gray')
plt.axhline(y=-2.58, color='gray')
```
