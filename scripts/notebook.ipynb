```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [20, 10]
```

```python
def read_db_from_dump():
    return pd.read_csv("data_dump")[["Project Name", "Project Language", "Technical Debt Ratio", "Number of Lines of Code"]]
```

```python
df = read_db_from_dump()
df = df.sort_values(by="Project Language")
```

```python
langz = list(df["Project Language"])
languages = []
for l in langz:
    if l not in languages:
        languages.append(l)
```

```python
languages
```

```python
import matplotlib.patches as mpatches

available_colors = ('b', 'r', 'g', 'y', 'k', 'm')
colors = []
for c in available_colors:
    colors = colors + [c] * 10

patches = []
for i, c in enumerate(available_colors):
    patches.append(mpatches.Patch(color=c, label=languages[i]))

ax = df.plot.bar(x="Project Name", y='Technical Debt Ratio', rot=90, color=colors)
ax.legend(handles=patches)
ax.get_figure().savefig('tdr.jpg')
```
