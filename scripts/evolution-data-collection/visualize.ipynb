```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [20, 10]
```

```python
df = pd.read_csv("data.csv")[
        ["Language", "Project", "Number of Lines of Code", "Technical Debt Ratio", "Technical Debt", "Size Normalized Technical Debt", "Date"]]
```

```python
# print(df.to_string())
```

```python
projects = [
    "ktor", 
    "express", 
    "dotty", 
    "rails", 
    "django", 
    "etcd",
    "finagle",
    "kotlinx.coroutines",
    "moby",
    "vue",
    "scikit-learn",
    "vagrant"
]
```

```python
for project in projects:
    project_df = df[df["Project"] == project]
    ax = project_df[["Size Normalized Technical Debt", "Date"]].plot(x="Date", grid=True, title=project + ": Evolution of Size-Normalized Technical Debt")
    ax = project_df[["Technical Debt", "Date"]].plot(x="Date", grid=True, title=project + ": Evolution of Technical Debt")
```
